```python
import pandas as pd
data=pd.read_csv('/content/10000 Sales Records.csv')
print("approch 4: imuted estimation missing values")
data_imputed_mean=data.fillna(data.mean(numeric_only=True))
print("imputed mean:")
print(data_imputed_mean)
print()
data_interpolated=data.interpolate(method='linear')
print("interpolated:")
print(data_interpolated)
print()

```

```python
import pandas as pd
data=pd.read_csv('/content/10000 Sales Records.csv')
z_score=(data-data.mean(numeric_only=True))/data.std(numeric_only=True)
outliers=z_score[abs(z_score)>3]
print("univariate outliers:")
print(outliers)
print()
```

```python
import pandas as pd
data=pd.read_csv('/content/10000 Sales Records.csv')
duplicate_rows=data[data.duplicated()]
print("duplicate rows:")
print(duplicate_rows)
data.drop_duplicates(inplace=True)
numeric_data=data.select_dtypes(include=['number'])
correlation_matrix = numeric_data.corr()
correlation_features = correlation_matrix[abs(correlation_matrix) > 0.5] # Filter correlations greater than 0.5 in absolute value
print("correlation features:")
print(correlation_features)
print()
```

```python
import pandas as pd
data=pd.read_csv('/content/10000 Sales Records.csv')
upper_cap=50
lower_cap=10
numeric_data=data.select_dtypes(include=['number'])
data_capped=numeric_data.clip(lower=lower_cap,upper=upper_cap)
print("capped data:")
print(data_capped)
print()
```
