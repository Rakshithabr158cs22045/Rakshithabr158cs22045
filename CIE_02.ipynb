```python
from scipy.spatial.distance import mahalanobis
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

path = ('/content/10000 Sales Records.csv')
df = pd.read_csv(path)
df=df[['Order ID','Country']]

mean = np.mean(df,axis=0)
cov_matrix=np.cov(df,rowvar=False)
inv_cov_matrix = np.linalg.inv(cov_matrix)
outliers=[]
for i in range(len(df)):
  point = df.iloc[i].values
  mahalanobis_dist = mahalanobis(point,mean,inv_cov_matrix)
  if mahalanobis_dist > -1:
    outliers.append(i)

df['outlier'] = False
df.loc[outliers, 'outlier'] = True

plt.figure(figsize=(12, 6))
plt.plot(df.index, df['Order ID'], label='Country', color='blue')

plt.scatter(df.index[df['outlier']], df['Order ID'][df['outlier']], color='red', label='Outliers')
plt.xlabel('Order ID')
plt.ylabel('Conuntry')

plt.show()
```
