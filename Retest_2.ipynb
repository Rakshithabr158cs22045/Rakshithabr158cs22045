```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data=pd.read_csv('/content/Costomers Sales(1).csv')
column1 = "Region"
column2 = "Country"
data['Region'] = pd.to_numeric(data['Region'], errors='coerce')
z_score1 = np.abs((data[column1] - data[column1].mean())/data[column1].std())
country_counts = data[column2].value_counts()
threshold = 10
outlier_countries = country_counts[country_counts < threshold].index.tolist()
outliers = data[(z_score1 > 3) | (data[column2].isin(outlier_countries))]
print(outliers)
plt.figure(figsize=(12, 6))
country_counts.plot(kind='bar')
plt.xlabel('Country')
plt.ylabel('Frequency')
plt.title('Country Frequencies')
plt.show()
```

```python
import pandas as pd
data=pd.read_csv('/content/Costomers Sales(1).csv')
duplicate_rows=data[data.duplicated()]
print("duplicate rows:")
print(duplicate_rows)
data.drop_duplicates(inplace=True)
numeric_data=data.select_dtypes(include=['number'])
correlation_matrix = numeric_data.corr()
correlation_features = correlation_matrix[abs(correlation_matrix) > 0.5] # Filter correlations greater than 0.5 in absolute value
print("correlation features:")
print(correlation_features)
print()
```
